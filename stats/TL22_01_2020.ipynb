```r
library(dplyr)
library(xts)
library(TTR)
```

```r
df =read.csv("../data/Daily_Demand_Forecasting_Orders.csv",sep =";")
head(df)
dim(df)
```

```r
str(df)
```

```r

orders  <- df[,c(1,2,13)]

names(orders)<- c("week","weekday","total")
```

```r
head(orders)
```

```r
order.ts<- ts(orders$total,start = 1,end =60,frequency = 5)
head(order.ts)
```

```r
plot(order.ts)
```

```r
lag3<- stats::lag(order.ts,3)
```

```r
ma5 <-TTR::SMA(order.ts,n=5)
ma3 <-TTR::SMA(order.ts,n=3)
```

```r
options(pepr.plot.width=60,repr.plot.height=3)
plot(order.ts)
lines(lag3,col="yellow")
lines(ma5,col="blue")
lines(ma3,col="green")
```

```r
stationaty_dts<-decompose(order.ts)
plot(stationaty_dts)
```
