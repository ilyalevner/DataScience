## Logistic Regression

```r
bd <- read.csv("../data/blood_donations.csv")
head(bd)
```

```r
summary(bd)
```

```r
table(bd$donate_2007)
```

```r
mod2 <- glm(donate_2007~.,data=bd,family="binomial")
summary(mod2)
```
